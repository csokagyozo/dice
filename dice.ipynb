```python
import numpy as np
import matplotlib.pyplot as plt


def generate_dist(d,k,n):
    dist = np.zeros(d*k)
    for i in range(n):
        sum = 0
        for j in range(k):
            sum += np.random.randint(d) + 1
        dist[sum-1] += 1
    return dist

```

```python
# n = 10
dist = generate_dist(6,2,10)
plt.bar(np.arange(len(dist)) + 1, dist)
plt.show()
```
